# Try and Except Block

```python
num = 20
try:
    den = int(input("Enter the value"))
    print(num/den)
except:
    print("An error occured")
#this print statment execute means that the flow does not break even error occur
print("Outside from try and except block")
```

## Catch an Error
when the try block throw an  error then we can catch that error in exception block

```python
num = 10
try:
    den = int(input("Enter the value"))
    print(num/den)
except Exception as e:
    print(e)
    
```

##### We can catch the specific exception

```python
# ValueError
num = 10
try:
    den = int(input("Enter the value"))
    print(num/den)
except ValueError:
    print("Provide integer value")
```

```python
# ZeroDivisionError
num = 10
try:
    den = int(input("Enter the value"))
    print(num/den)
except ZeroDivisionError:
    print("Don't give denominator as 0")
```

## We can catch multiple errors. But execute in order wise

```python
num = 10
try:
    den = int(input("Enter the value"))
    print(num/den)
except ZeroDivisionError:
    print("Don't give denominator as 0")
except ValueError:
    print("Provide integer value")
except Exception as e:
    print("Got a error")
```

### try-except-else
- when try does not throw an error theen, the else block gets executed.
- It is good place to put code that should run if no exception are raised

```python
num = 10
try:
    den = int(input("Enter the value"))
    print(num/den)
except ZeroDivisionError:
    print("Don't give denominator as 0")
except ValueError:
    print("Provide integer value")
except Exception as e:
    print("Got a error")
else:
    print("Great, You have given good number")
```

## try - except - else - finally
- the finally block is always execute whether an exception is raised or not

```python
num = 10
try:
    den = int(input("Enter the value"))
    print(num/den)
except ZeroDivisionError:
    print("Don't give denominator as 0")
except ValueError:
    print("Provide integer value")
except Exception as e:
    print("Got a error")
else:
    print("Great, You have given good number")
finally:
    print("This block is always execute")
```

# Create a Custom Error

```python
#class for error
class WrongEmailError(Exception):
    pass
```

```python
class Person:
    def __init__(self,salary,email):
        if salary > 1000000:
            raise ValueError("Enter valid salary")
        self.salary = salary
        if ('@' not in email):
            raise WrongEmailError("Enter valid email")
        self.email = email
    def show_salary(self):
        print(self.salary)
    def show_email(self):
        print(self.email)
```

```python
try:
    p1 = Person(10000,"k1010")
except WrongEmailError as  e:
    print(e)
```

```python
p1.show_salary(),p1.show_email()
```
