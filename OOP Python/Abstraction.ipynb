# Abstraction
Abstraction is used to hide the implementation details from the user and expose only necessary parts

```python
from abc import ABC, abstractmethod
```

```python
class user(ABC):
    def __init__(self):
        pass
    @abstractmethod   
    def login(self):
        pass
    def auth(self):
        pass
    def logut(self):
        print("Logged out")
        
```

```python
class emp(user):
    def login(self):
        print("log in")
    def info(self):
        pass
```

```python
em1 = emp()
```

```python
em1.logut()
```
