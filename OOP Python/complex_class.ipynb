```python
# Complex class by own
class ComplexNumber():
    def __init__(self,real=0.0,imag=0.0):
        self.real = real
        self.imag = imag

    # __str__ magic method : this is called when we use print funciton
    def __str__(self):
        if self.imag == 0:
            imag = int(self.imag)
            return f'{imag}i'
        elif self.imag<0:
            com_num = (f'({self.real}{self.imag}i)')
        else:
            com_num = (f'({self.real}+{self.imag}i)')
        return com_num

    # method for creating conjugate number    
    def conjugate(self):
        return ComplexNumber(self.real,-1*self.imag)


    # addtion method for complex number
    def __add__(self,other):

        real = self.real + other.real
        imag = self.imag + other.imag
        return ComplexNumber(real,imag)


    def __sub__(self,other):

        real = self.real - other.real
        imag = self.imag - other.imag
        return ComplexNumber(real,imag)

    # method for multiplication
    def __mul__(self,other):
        real = self.real*other.real-self.imag*other.imag
        imag = self.real*other.imag+self.imag*other.real
        return ComplexNumber(real,imag)

    # method for devision
    def __truediv__(self,other):

        den = other.real**2 + other.imag**2
        ans = self*ComplexNumber(other.real/den, (-1*other.imag)/den)
        return ans
        
```

```python
cn2=ComplexNumber(3,4)
cn1 = ComplexNumber(3,6)
```

```python
cn = ComplexNumber(8,13)
```

```python
print(f"Conjugate of {cn2}",cn2.conjugate())

print(f"addition of {cn}, {cn2} and {cn1} :",cn+cn2+cn1)

print(f"subtraction {cn} , {cn2} and {cn1} :",cn-cn2-cn1)
print(f"subtraction {cn2} and {cn} :",cn2-cn)

print(f"Multiplication of {cn2} and {cn1} :",cn2*cn1)
print(f"division of {cn2} and {cn1} :",cn2/cn1)
```
