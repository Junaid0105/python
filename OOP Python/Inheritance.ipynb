# Inheritance

```python
#Parent class
class user:
    def __init__(self, name, rollNumber, marks):
        self.name = name
        self.rollNumber = rollNumber
        self.marks = marks
    def login(self):
        print("Logged in")

    def loggout(self):
        print("Logged out")

```

```python
#child class
class student(user):
    def __init__(self,name,rollNumber,marks,id,age):
        self.id = id
        self.age = age
        
        # we use super() keywords to access methodes of parents class
        # we cannot use outside the class
        super().__init__(name,rollNumber,marks)
        
    def info(self):
        print("Name : ", self.name)
        print("Roll Number : ", self.rollNumber)
```

```python
s1 = student("Junaid",1,99,100,25)
```
