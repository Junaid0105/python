# Simple Inheritance

```python
class grandfather:
    def __init__(self,name, age):
        self.name = name
        self.age = age

    def show_details(self):
        return f'Name: {self.name}, Age: {self.age}'

    def speak(self):
        return 'grandfather speak wisely'

class father(grandfather):
    def __init__(self, name, age, occupation):
        super().__init__(name, age)
        self.occupation = occupation
    def show_occupation(self):
        return f'Occupation: {self.occupation}'

    def speak(self):
        return 'Father speak carefully'

ather_obj = father("Naiem",50,"Farmer")
print(ather_obj.show_details())
print(ather_obj.show_occupation())
print(ather_obj.speak())
```

# Multilevel

```python
class grandfather:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def show_detials(self):
        return f'Name: {self.name} , Age: {self.age}'
    def speak(self):
        return "Grandfather speak wisely"

class father(grandfather):
    def __init__(self,name, age, occupation):
        super().__init__(name,age)
        self.occupation = occupation
    def show_occupation(self):
        return f'Occupation: {self.occupation}'
    def speak(self):
        return 'father speak carefully'
class child(father):
    def __init__(self, name,age,hobby):
        super().__init__(name, age, hobby)
        self.hobby = hobby
    def show_hobby(self):
        return f'Hobby: {self.hobby}'
```

```python
# objects of child class
child_obj = child("Junaid",25,"Playing Cricket")
print(child_obj.show_detials())
print(child_obj.show_occupation())
print(child_obj.show_hobby())
print(child_obj.speak())
```

# Multiple

```python
class father:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def show_details(self):
        return f'Name: {self.name}, Age: {self.age}'
    def speak(self):
        return 'Father speak carefully'
class mother:
    def __init__(self, name, favorite_food):
        self.name = name
        self.favorite_food = favorite_food
    def show_favorite_food(self):
        return f'Favorite Food: {self.favorite_food}'
    def speak(self):
        return 'Mother speak lovingly'
class child(father, mother):
    def __init__(self, name, age, favorite_food, hobby):
        father.__init__(self, name, age)
        mother.__init__(self,name, favorite_food)
        self.hobby = hobby
    def show_hobby(self):
        return f'Hobby: {self.hobby}'
    
```

```python
# object for child class
child_obj = child("Junaid",25,"Biryani","Playing Cricket")
```

```python
print(child_obj.show_details())
print(child_obj.show_favorite_food())
print(child_obj.show_hobby())
print(child_obj.speak())
```
