```python
#create parent class of the institute as Cadd
class Cadd:
    def __init__(self,course=None,course_id=None):
        self.course = course
        self.course_id = course_id
        
    def course_info(self):
        print(f"Course ID: {self.course_id} | Course Name: {self.course}")
        
```

```python
cd = Cadd()
```

```python
#create subclass (child class) as faculty
class student(Cadd):
    def __init__(self,st_name,trainer_name,course,course_id):
        self.st_name = st_name
        self.trainer_name = trainer_name
        super().__init__(course,course_id)
    def student_info(self):
        print("Student Name: ",self.st_name)
        print("Course Name: ", self.course)
        print("Course ID: ", self.course_id)
        print("Trainer name: ", self.trainer_name)
        
        
```

```python
st1 = student("Junaid","ALAM","Data",9)
```

```python
st1.student_info()
```
