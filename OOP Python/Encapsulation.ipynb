```python
class Students:

    def __init__(self, name, rollNumber, marks):
        self.__name = name ## private name variable
        self.__rollNumber = rollNumber ## private rollnumber
        self.__marks = marks ## private marks

    # methods for accessing private variables value
    # getter method
    def getMarks(self):
        return self.__marks
    def getName(self):
        return self.__name

    # methods for updating the private variables value
    #setter method
    def setMarks(self,NewMarks,passcode):
        if passcode == "0000": # security
            self.__marks = NewMarks
        else:
            print("Wrong passcode")
    def study(self):
        print("I am studing")

    def play(self):
        print("I am playing")

    
```

```python
s1 = Students("Junaid",1,90)
s2 = Students("Alam",2,85)
```

```python
print(s1.getMarks())
print(s2.getMarks())
```

```python
s1.setMarks(56,"0000")
```

```python
class Students:
    __NumberOfStudent = 0
    __SchoolName = 'RRK'
    name = "temp"

    def __init__(self, name, rollNumber, marks):
        self.__name = name ## private name variable
        self.__rollNumber = rollNumber ## private rollnumber
        self.__marks = marks ## private marks

    # methods for accessing private variables value
    # getter method
    def getMarks(self):
        return self.__marks
    def getName(self):
        return self.__name

    # methods for updating the private variables value
    #setter method
    def setMarks(self,NewMarks,passcode):
        if passcode == "0000": # security
            self.__marks = NewMarks
        else:
            print("Wrong passcode")
    #this is decorator
    @staticmethod
    def getSchoolName():
        return Students.__SchoolName
    @staticmethod
    def setSchoolName(NewSchool):
        Students.__SchoolName = NewSchool
    def study(self):
        print("I am studing")

    def play(self):
        print("I am playing")
```
