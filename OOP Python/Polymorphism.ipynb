Polymorphism in python is refers to ability of the same method or operation behave different based on object or context

# Overloading
Python does not suppert method overloading by default

```python
#Python does not suppert overloading by default
class user:
    #constructor 1
    def __init__(self,name,age):
        self.name = name
        self.age = age
    # constructor 2 ----> construtor 2 overwrite constructor 1
    def __init__(self, name,age,place):
        self.name = name
        self.age = age
        self.place = place

#create user object
u1 = user("Junaid",25,"Amroha")
print(u1.age,u1.name,u1.place)
u2 = user("Alam",34) ## this object give error because it expect 3 arguments
    
```

### We use different techniques to achieve overloading
1. Using Variables Arguments (*args)
2. Using Default Arguments (None as default values)

```python
class user:
    def __init__(self,name=None,age=None,place=None):
        if (place == None):
            self.c1(name,age)
        else:
            self.c2(name,age,place)
    def c1(self,name,age):
        self.name = name
        self.age = age
    def c2(self,name,age,place):
        self.name = name
        self.age = age
        self.place = place
```

```python
u1 = user("Junaid",25)
print(u1.name,u1.age)
```

```python
u2 = user("Junaid",24,"Pakbara")
print(u2.name,u2.age,u2.place)
```

# Overriding

```python
class Animal:
    def sound(self):
        return "Some generic sound"
class Dog(Animal):
    def sound(self):
        return 'Bark'
d = Dog()
print(d.sound())
a = Animal()
print(a.sound())
```
